```python
import pandas as pd

# B·∫£ng ƒëi·ªÉm chu·∫©n
benchmarks = {
    "To√°n CH": 30.7,
    "VƒÉn CH": 28.3,
    "Anh CH": 28.2,
    "Tin CH": 22.5,
    "H√≥a CH": 33,
    "Sinh CH": 28.6
}


# ƒê·∫£m b·∫£o t·∫•t c·∫£ c√°c ƒëi·ªÉm l√† s·ªë th·ª±c (float), thay 'null' b·∫±ng 0
cols_to_convert = ["To√°n KC", "VƒÉn KC", "Anh KC"] + list(benchmarks.keys())
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors="coerce").fillna(0)

# X√°c ƒë·ªãnh c√°c m√¥n chuy√™n
specialized_subjects = list(benchmarks.keys())

# T·∫°o danh s√°ch m·ªõi l∆∞u d·ªØ li·ªáu theo c·ªôt
data_list_specialized_subjects = []

for index, row in df.iterrows():
    specialized_taken = [sub for sub in specialized_subjects if row[sub] > 0]

    # T·∫°o dictionary ƒë·ªÉ l∆∞u d·ªØ li·ªáu c·ªßa th√≠ sinh
    candidate_data = {
        "S·ªë b√°o danh": row["S·ªë b√°o danh"],
    }

    # Ghi d·ªØ li·ªáu m√¥n chuy√™n v√†o c·ªôt "M√¥n chuy√™n 1", "M√¥n chuy√™n 2", ...
    for i in range(2):
        if i < len(specialized_taken):
            sub = specialized_taken[i]
            candidate_data[f"Specialized Subjects {i+1}"] = sub
            candidate_data[f"Specialized Score {i+1}"] = row[sub]
            candidate_data[f"Specialized Score x 2 {i+1}"] = row[sub] * 2 
        else:
            candidate_data[f"Specialized Subjects {i+1}"] = "null"
            candidate_data[f"Specialized Score {i+1}"] = 0
            candidate_data[f"Specialized Score x 2 {i+1}"] = 0

    data_list_specialized_subjects.append(candidate_data)

# T·∫°o DataFrame t·ª´ danh s√°ch d·ªØ li·ªáu
df_specialized = pd.DataFrame(data_list_specialized_subjects)
```

```python
pass_list = []
fail_list = []

for index, row in df.iterrows():
    # Ki·ªÉm tra n·∫øu 3 m√¥n ch√≠nh b·ªã thi·∫øu ho·∫∑c t·∫•t c·∫£ m√¥n chuy√™n b·ªã thi·∫øu
    if row["To√°n KC"] == 0 or row["VƒÉn KC"] == 0 or row["Anh KC"] == 0 or all(row[sub] == 0 for sub in specialized_subjects):
        fail_list.append({**row, "Status": "Failed", "Remarks": "Not enough information", "Results": ""})
    else:
        base_score = row["To√°n KC"] + row["VƒÉn KC"] + row["Anh KC"]
        specialized_taken = [sub for sub in specialized_subjects if row[sub] > 0]

        result_details = []
        passed_subjects = []
        failed_subjects = []

        # T√≠nh ƒëi·ªÉm cho t·ª´ng m√¥n chuy√™n
        for sub in specialized_taken:
            specialized_score_1 = df_specialized.loc[index, "Specialized Score x 2 1"]  # ƒêi·ªÉm c·ªôt 1
            specialized_score_2 = df_specialized.loc[index, "Specialized Score x 2 2"]  # ƒêi·ªÉm c·ªôt 2

            total_score_1 = base_score + specialized_score_1  # T·ªïng ƒëi·ªÉm v·ªõi c·ªôt 1
            total_score_2 = base_score + specialized_score_2  # T·ªïng ƒëi·ªÉm v·ªõi c·ªôt 2

            # L∆∞u k·∫øt qu·∫£ v√†o danh s√°ch theo format y√™u c·∫ßu
            if specialized_score_1 > 0:
                result_details.append(f"{sub}: {total_score_1}")
            if specialized_score_2 > 0:  # N·∫øu c·ªôt 2 kh√¥ng ph·∫£i 0 th√¨ m·ªõi ghi
                result_details.append(f"{sub}: {total_score_2}")

            # Ki·ªÉm tra n·∫øu √≠t nh·∫•t m·ªôt t·ªï h·ª£p ƒë·∫°t chu·∫©n
            if (specialized_score_1 > 0 and total_score_1 >= benchmarks.get(sub, float('inf'))) or \
               (specialized_score_2 > 0 and total_score_2 >= benchmarks.get(sub, float('inf'))):
                passed_subjects.append(sub)
            else:
                failed_subjects.append(sub)

        # G·ªôp c√°c k·∫øt qu·∫£ th√†nh chu·ªói
        result_string = "; ".join(result_details)

        # N·∫øu c√≥ √≠t nh·∫•t m·ªôt m√¥n ƒë·∫°t chu·∫©n, th√¨ Pass
        if passed_subjects:
            pass_list.append({**row, "Status": "Pass", "Remarks": "; ".join(passed_subjects), "Results": result_string})

        # N·∫øu c√≥ √≠t nh·∫•t m·ªôt m√¥n kh√¥ng ƒë·∫°t chu·∫©n, th√¨ c≈©ng l∆∞u v√†o danh s√°ch Fail
        if failed_subjects:
            fail_list.append({**row, "Status": "Failed", "Remarks": "; ".join(failed_subjects), "Results": result_string})

# T·∫°o DataFrame t·ª´ danh s√°ch Pass v√† Fail
df_pass = pd.DataFrame(pass_list)
df_fail = pd.DataFrame(fail_list)
df_pass["S·ªë b√°o danh"] = df_pass["S·ªë b√°o danh"].astype(int)
df_fail["S·ªë b√°o danh"] = df_fail["S·ªë b√°o danh"].astype(int)
```

```python
df_pass.to_excel("Pass_Report.xlsx", index=False)
df_fail.to_excel("Failed_Report.xlsx", index=False)
```
